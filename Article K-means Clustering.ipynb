```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
# Generate synthetic customer data
data, _ = make_blobs(n_samples=500, centers=5, n_features=2, random_state=42)
```

```python
# Visualize the synthetic customer data
plt.scatter(data[:, 0], data[:, 1])
plt.title("Customer Segmentation Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Cluster the customers into different segments
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(data)
```

```python
# Visualize the clustered customer segments
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='r')
plt.title("Customer Segmentation Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Analyze the customer segments
segment_labels = kmeans.predict(data)
segment_data = pd.DataFrame(data, columns=["Feature 1", "Feature 2"])
segment_data["Segment"] = segment_labels
segment_summary = segment_data.groupby("Segment").mean()
print(segment_summary)
```
